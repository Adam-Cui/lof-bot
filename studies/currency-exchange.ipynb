```python
inv = get_daily("currencies/usd-cny", start="20180101")
```

```python
zjj = get_daily("USD/CNY", start="20180101")
```

```python
inv = inv[inv["date"].isin(xa.cons.opendate)]
```

```python
cmp = zjj.merge(inv, on=["date","date"], how="inner", suffixes=["_zjj", "_inv"])
```

```python
dev = (cmp["close_zjj"]-cmp["close_inv"].shift(1))/cmp["close_zjj"]
```

```python
np.mean(dev) # 市场价更看贬人民币（当然了）
```

```python
np.mean(np.abs(dev)) ## 偏差在千一六，中间价和前一日收盘价类比更接近
```

```python
dev = (cmp["close_zjj"]-cmp["open"])/cmp["close_zjj"]
```

```python
np.mean(np.abs(dev)) ## 和当日开盘差别也不大
```

```python
cmp["percent_zjj"] = ((cmp["close_zjj"]/cmp["close_zjj"].shift(1)-1)*100).shift(-1)
cmp["percent_inv"] = cmp["percent"].apply(lambda x: float(x[:-1]))
```

```python
cmp.iloc[-30:] # 注意 T 日中间价早晨报价基本上反应的是前一日的市价收盘价
```

```python
cmp["percent_zjj"].corr(cmp["percent_inv"]) # 不够大啊，尴尬==, 但总之把 T 日中间价和 T-1 日市场收盘价关联是最靠谱的了
```
